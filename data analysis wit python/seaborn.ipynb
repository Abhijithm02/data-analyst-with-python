Data Visualisation with Seaborn

Seaborn is a python visualization library based on matplotlib that provides a high-level interface for drawing attractive and informative stastistical graphics seaborn helps in creating complex visualization with just a few lines of code

```python
import seaborn as sns
```

```python
# basic ploting with seaborn
tips=sns.load_dataset('tips')
tips
```

```python
##create a scatter plot
import matplotlib.pyplot as plt
tips=sns.load_dataset('tips')
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("scatter plots of total bill vs tip")
plt.show()
```

```python
## line plot

sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("line plot of total bill by size")
plt.show()
```

```python
## categorical plot
##bar plot
sns.barplot(x='day',y='total_bill',data=tips)
plt.title("bar plot of total bill by day")
plt.show()
```

```python
## box plot
sns.boxplot(x="day",y='total_bill',data=tips)
```

```python
## violin plot

sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
## histograms
sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
## kde plot
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
## pairplot
sns.pairplot(tips)
```

```python
## heatmap
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
