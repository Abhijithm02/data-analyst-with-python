Pandas-Dataframe and series


pandas is a powerful data manipulation library ipython widely used for data anaysis and data cleaning.it provide two primary data structures:seris and dataframes.a series is a one dimensional array-like object,while a dataframe is a two-dimentional, size-mutable, and potentialy heterogeneous tabular data structure with a labeled axes (rows and coulmns).

```python
##series
import pandas as pd
data=[1,2,3,4,5]
series=pd.Series(data)
print("series \n",series)
```

```python
## create a sereis from dictionary
import pandas as pd
data={'a':1, 'b':2,'c':3}
series_dict=pd.Series(data)
print(series_dict)
```

```python
data=[10,20,30]
index=['a','b','c']
pd.Series(data,index=index)
```

```python
#dataframe
#create a data frame from a dictionary of list

data={
    'Name':['krish','jhon','jack'],
    'Age':[25,30,45],
    'City':['Banglore','New york','florida']
}
df=pd.DataFrame(data)
print(df)
print(type(df))
```

```python
import numpy as np
np.array(df)
```

```python
##create a dataframe from a list of dictionaries
data=[
    {'Name':'krish','Age':32,'City':'bangalore'},
    {'Name':'end','Age':34,'City':'bangalore'},
    {'Name':'jack','Age':45,'City':'bangalore'},
     {'Name':'dennis','Age':32,'City':'bangalore'}
]
df=pd.DataFrame(data)
print(df)
print(type(df))
```
